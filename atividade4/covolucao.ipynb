```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
%matplotlib inline
```

```python
n = np.arange(-5,6,1)
def x(n): return np.where( np.abs(n)>2 , 0,  n**2 + 5)
```

#### Sinal discreto x[n]:

```python
fig, (ax1) = plt.subplots(1, 1, figsize=(6, 3)) #criação do subplot
ax1.stem(n,x(n), label="$x(n)$") #sinal continuo
ax1.set_xticks(n) #valores do eixo x
ax1.set_xlabel('$n$') #rótulo do eixo x
ax1.legend()
plt.grid()
```

#### Sistema S[n]:

```python
def s(n): return 2*n
```

```python
fig, (ax1) = plt.subplots(1, 1, figsize=(6, 3)) #criação do subplot
ax1.stem(n,s(x(n)), label="$S(n)$") #sinal continuo
ax1.set_xticks(n) #valores do eixo x
ax1.set_xlabel('$n$') #rótulo do eixo x
ax1.legend()
plt.grid()
```

```python
def d(n): return np.where( n==1 , 1, 0)
```

```python
fig, (ax1) = plt.subplots(1, 1, figsize=(6, 3)) #criação do subplot
ax1.stem(n,d(n), label="$d(n)$") #sinal continuo
ax1.set_xticks(n) #valores do eixo x
ax1.set_xlabel('$n$') #rótulo do eixo x
ax1.legend()
plt.grid()
```
