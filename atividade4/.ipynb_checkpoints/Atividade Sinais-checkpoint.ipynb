```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

<div class="alert alert-info" style="color:#1a96f6;">
1ª TAREFA:Implementação do algorítimo do cálculo da convolução
</div>

```python
def convolution(h, x):
    len_h = len(h)
    len_x = len(x)
    
    if(len_h == len_x):
        result = np.zeros(len(h) + len(x) - 1)
    elif(len_h > len_x):
        dife = len_h - len_x
        for i in range(dife):
            x.append(0)
            
        result = np.zeros(len(h) + len(x) - 1)
    else:
        dife = len_x - len_h
        for i in range(dife):
            h.append(0)
            
        result = np.zeros(len(h) + len(x) - 1)

    for i in reversed(range(len(x))):
        for j in range(len(h)):
            result[i + j] += h[i] * x[j]
            
    if(len_h != len_x):
        for i in range(dife):
            result = np.delete(result,-1)

    return result
```

<div class="alert alert-info" style="color:#1a96f6;">
2ª TAREFA:Testes com exemplos da lista
</div>

```python
h = [1, 2, 1]
x = [2, 3, -2]

convolution(h, x)
```

```python
h = [2, 0, 2]
x = [1, 2, 0 ,-1]

convolution(h, x)
```

<div class="alert alert-info" style="color:#1a96f6;">
3ª TAREFA: obter os mesmos resultados do notebook do link
2
</div>

```python
def convolution(h, x):
    return np.convolve(h, x, mode='full')

# Definindo h[n] e x[n]
h = np.array([1, 2, 1])
x = np.array([2, 3, -2, 0])

# Calculando a convolução
result = convolution(h, x)

# Plotando os sinais h[n] e x[n]
plt.stem(range(len(h)), h, label='h[n]', basefmt='b-')
plt.stem(range(len(x)), x, label='x[n]', basefmt='r-')
plt.legend()
plt.title('Sinais de entrada h[n] e x[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.show()

# Plotando o resultado da convolução
plt.stem(range(len(result)), result, label='y[n]', basefmt='g-')
plt.legend()
plt.title('Resultado da Convolução y[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.show()
```

```python
def convolution(h, x):
    return np.convolve(h, x, mode='full')

# Definindo h[n] e x[n]
h = np.array([0, 2, 0])
x = np.array([1, 2, 0, -1])

# Calculando a convolução
result = convolution(h, x)

# Plotando os sinais h[n] e x[n]
plt.stem(range(len(h)), h, label='h[n]', basefmt='b-')
plt.stem(range(len(x)), x, label='x[n]', basefmt='r-')
plt.legend()
plt.title('Sinais de entrada h[n] e x[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.show()

# Plotando o resultado da convolução
plt.stem(range(len(result)), result, label='y[n]', basefmt='g-')
plt.legend()
plt.title('Resultado da Convolução y[n]')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.show()
```
