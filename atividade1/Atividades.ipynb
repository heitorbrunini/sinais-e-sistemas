<div class="alert alert-info" style="color:#1a96f6;">
2. Considere o sinal $y(t)$ definido a seguir, e obtenha o resultado de transformação solicitada.

\begin{equation}
y(t)=
\begin{cases} 
\left(\frac{1}{5}\right)x(-2t-3) , \text{para } 0 \leq t \leq 2 \\[2mm]
0 , \text{caso contrário} 
\end{cases}
\end{equation}
</div>

Função:

```python
def s(t): return np.where( np.abs(t)>2 , 0, (-1 *(2*t + 3) ) * t / 5) 
```

Valores:

```python
t = np.arange(-4,10,1)
n = np.arange(-4,10)
```

Sinal Original:

```python
fig, (ax1) = plt.subplots(1, 1, figsize=(10, 4)) #criação do subplot
ax1.plot(t,s(t), label="$y(t)$") #sinal continuo
ax1.set_xticks(n) #valores do eixo x
ax1.set_xlabel('$t$') #rótulo do eixo x
ax1.legend()
```

<div style="color:#1a96f6;">
a) $y(1 - 3t)$
</div>

Aceleração e deformação do Sinal

```python
fig, (ax1) = plt.subplots(1, 1, figsize=(10, 4)) #criação do subplot
ax1.plot(t,s( (t*3) -1 ), label="$y(1 - 3t)$") #sinal continuo
ax1.set_xticks(n) #valores do eixo x
ax1.set_xlabel('$t$') #rótulo do eixo x
ax1.legend()
```

<div style="color:#1a96f6;">
b) $y(1 - t) + y(2 - t)$
</div>

```python
aux = s(1 - t) + s(2 - t)
fig, (ax1) = plt.subplots(1, 1, figsize=(10, 4)) #criação do subplot
ax1.plot(t,aux, label="$y(1 - t) + y(2 - t)$") #sinal continuo
ax1.set_xticks(n) #valores do eixo x
ax1.set_xlabel('$t$') #rótulo do eixo x
ax1.legend()
```

<div style="color:#1a96f6;">
c) $y(1 - t) \cdot y(2 - t)$
</div>

```python
aux = s(1 - t) * s(2 - t)
fig, (ax1) = plt.subplots(1, 1, figsize=(10, 4)) #criação do subplot
ax1.plot(t,aux, label="$y(1 - t) + y(2 - t)$") #sinal continuo
ax1.set_xticks(n) #valores do eixo x
ax1.set_xlabel('$t$') #rótulo do eixo x
ax1.legend()
```

<div style="color:#1a96f6;">
d) $y\left(4 - \dfrac{t}{2}\right)$
</div>

```python
aux = s(4 - t/2)
fig, (ax1) = plt.subplots(1, 1, figsize=(10, 4)) #criação do subplot
ax1.plot(t,aux, label="$ y(4 - t/2)$") #sinal continuo
ax1.set_xticks(n) #valores do eixo x
ax1.set_xlabel('$t$') #rótulo do eixo x
ax1.legend()
```

<div style="color:#1a96f6;">
e) $y\left[\dfrac{1}{2}(t - 1)^2\right]$
</div>

```python
aux = s(  ((t-1)**2)/2  )
fig, (ax1) = plt.subplots(1, 1, figsize=(10, 4)) #criação do subplot
ax1.plot(t,aux, label="$ s( (t-1)**2 ) /2)$") #sinal continuo
ax1.set_xticks(n) #valores do eixo x
ax1.set_xlabel('$t$') #rótulo do eixo x
ax1.legend()
```
